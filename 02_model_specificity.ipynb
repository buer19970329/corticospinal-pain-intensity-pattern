```python
import joblib
import numpy as np
import pandas as pd
from scipy.stats import pearsonr
```

# study 2 low itch/ high itch

```python
qh_li_copes = np.load('0data/d2_qh_li_set.npy')
qh_hi_copes = np.load('0data/d2_qh_hi_set.npy')
qh_li_y = np.load('0data/d2_qh_li_y.npy')
qh_hi_y = np.load('0data/d2_qh_hi_y.npy')
lassopcr = joblib.load('0data/lassopcr.pkl')
```

```python
print(pearsonr(qh_li_y, lassopcr.predict(qh_li_copes)))
print(pearsonr(qh_hi_y, lassopcr.predict(qh_hi_copes)))
```

# study 3 empathy

```python
empathy_pain_data = np.load('0data/d3_empathy_pain_set.npy')
empathy_nopain_data = np.load('0data/d3_empathy_nopain_set.npy')
empathy_rating = pd.read_csv('0data/d3_empathy_rating.csv')
```

```python
print(pearsonr(lassopcr.predict(empathy_pain_data), empathy_rating['PR'].values))
print(pearsonr(lassopcr.predict(empathy_nopain_data), empathy_rating['NPR'].values))
```
